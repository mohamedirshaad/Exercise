```python
data["ratings"]=data["ratings"].fillna("0")
```

```python
data[["no_of_ratings","discount_price","actual_price"]]=data[["no_of_ratings","discount_price","actual_price"]].fillna("0")
```

```python
data.isna().sum()
```

```python
data.loc[data["ratings"]=='Get',:]
```

```python
data.loc[data["ratings"]=='Get',"ratings"]="0.0"
```

```python
data["actual_price"]=data["actual_price"].str.replace(",","")
```

```python
data["actual_price"]=data["actual_price"].str.replace("₹","")
data["discount_price"]=data["discount_price"].str.replace("₹","")
data["discount_price"]=data["discount_price"].str.replace(",","")
data["no_of_ratings"]=data["no_of_ratings"].str.replace(",","")
```

```python
data.head()
```

```python
data["discount_price"]=data["discount_price"].astype(float)
data["actual_price"]=data["actual_price"].astype(float)
data["ratings"]=data["ratings"].astype(float)
```

```python
data.info()
```

```python
data["no_of_ratings"].unique()
```

```python
data.loc[data["no_of_ratings"]=='Only 2 left in stock.',"no_of_ratings"]="0"
```

```python
data.loc[data["no_of_ratings"]=='Only 1 left in stock.',"no_of_ratings"]="0"
```

```python
data["no_of_ratings"]=data["no_of_ratings"].astype(int)
```

```python
data.info()
```

```python
data.loc[data["discount_price"]==0,"discount_price"]=avg_discount_prize
```
